## Change data format

```python
import pandas as pd

data_df = pd.read_csv("data/corporate_relocation.csv")
for i in range(len(data_df)):
    for filed in ["是否重点企业", "是否走访", "是否领导挂点企业", "承诺不搬离"]:
        if data_df[filed][i] == "是":
            data_df[filed][i] = 1
        elif data_df[filed][i] == "否":
            data_df[filed][i] = 0
data_df.to_csv("data/corporate_relocation.csv", index=False)
```

## Compare execution results between predicted sql and gpt-4 generated sql

```python
import pandas as pd

sql_df = pd.read_csv("./data/test_data-gpt4.csv")
sql_df.head()
```

```python
import pymysql.cursors

def get_sql_execution_result(sql):
    connection = pymysql.connect(
        host='localhost',
        port=3306,
        user='user',
        password='MySQL123456!',
        database='linewell',
        cursorclass=pymysql.cursors.DictCursor
    )
    with connection:
        with connection.cursor() as cursor:
            try:
                cursor.execute(sql)
                result = cursor.fetchall()
            except:
                result = "SQL语句错误"
    return result
```

```python
from tqdm import tqdm

sql_df["sql-预期执行结果"] = ""
sql_df["sql-gpt4执行结果"] = ""
for i in tqdm(range(len(sql_df))):
    row = sql_df.iloc[i]
    sql_gpt4 = row["sql-gpt4"]
    if sql_gpt4.startswith("SQLQuery:"):
        sql_gpt4 = sql_gpt4.replace("SQLQuery:", "").strip(' "')
        # print(sql_gpt4)
    sql_df["sql-gpt4执行结果"][i] = get_sql_execution_result(sql_gpt4)
    sql_df["sql-预期执行结果"][i] = get_sql_execution_result(row["预期sql"])
new_sql_df = sql_df[["question", "预期sql", "sql-预期执行结果", "sql-gpt4", "sql-gpt4执行结果", ]]
new_sql_df.to_csv("./data/test_data-gpt4-execution.csv", index=False)
```
